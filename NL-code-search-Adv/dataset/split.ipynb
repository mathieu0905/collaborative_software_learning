```python
import json

jsonl_file_path = 'train.jsonl'

repo_counts = {}

with open(jsonl_file_path, 'r') as file:
    for line in file:
        json_object = json.loads(line)
        
        if "repo" in json_object:
            repo_value = json_object
            if repo_value in repo_counts:
                repo_counts[repo_value] += 1
            else:
                repo_counts[repo_value] = 1

sorted_repo_counts = dict(sorted(repo_counts.items(), key=lambda x: x[0]))

for repo, count in sorted_repo_counts.items():
    print(f"repo: {repo}, count: {count}")
```

```python
import json

input_jsonl_file = 'sampled_train.jsonl'

output_jsonl_file = 'sampled_train.jsonl'

repo_counts = {}

with open(input_jsonl_file, 'r') as input_file:
    for line in input_file:
        json_object = json.loads(line)
        if "repo" in json_object:
            repo_value = json_object["repo"]
            
            if repo_value in repo_counts:
                repo_counts[repo_value].append(json_object)
            else:
                repo_counts[repo_value] = [json_object]

sorted_repo_names = sorted(repo_counts.keys())

with open(output_jsonl_file, 'w') as output_file:
    for repo_name in sorted_repo_names:
        for json_object in repo_counts[repo_name]:
            json.dump(json_object, output_file)
            output_file.write('\n')
```

```python
import json
import os

input_jsonl_file = 'sampled_train.jsonl'

output_folder = 'split_2_jsonl_files'

os.makedirs(output_folder, exist_ok=True)
num = 0
os.makedirs(output_folder, exist_ok=True)
user_objects = {}
with open(input_jsonl_file, 'r') as input_file:
    for line in input_file:
        num = num + 1
        json_object = json.loads(line)
        repo_value = json_object.get("repo", "")
        username = repo_value.split('/')[0]
        if username not in user_objects:
            user_objects[username] = []
        user_objects[username].append(json_object)
jsons_per_subset = num // 2
current_subset = []
file_number = 0
print(f"JSONs per Subset: {jsons_per_subset}")
for username, user_jsons in user_objects.items():
    current_subset.extend(user_jsons)
    if len(current_subset) >= jsons_per_subset:
        output_file_path = os.path.join(output_folder, f'output_{file_number}.jsonl')
        with open(output_file_path, 'w') as output_file:
            for json_object in current_subset:
                json.dump(json_object, output_file)
                output_file.write('\n')
        print(f"Subset {file_number}: {len(current_subset)} JSONs")
        current_subset = []
        file_number += 1
if current_subset:
    output_file_path = os.path.join(output_folder, f'output_{file_number}.jsonl')
    with open(output_file_path, 'w') as output_file:
        for json_object in current_subset:
            json.dump(json_object, output_file)
            output_file.write('\n')
    
    print(f"Subset {file_number}: {len(current_subset)} JSONs")
```
